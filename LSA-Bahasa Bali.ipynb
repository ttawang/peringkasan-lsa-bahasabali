```python
import pandas as pd
from sklearn.decomposition import TruncatedSVD
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD

#import library yg dibutuhkan
```

```python
data = [] #import data mentah dari folder
for i in range(150):
    f = open("data/mentah/"+str(i+1)+".txt", "r+", encoding='utf-8')
    lines = f.readlines()
    del lines[0]
    f.close()
    a = " ".join(lines)
    data.append(a.replace("\n", " "))
    
#memasukkan data kedalam array untuk nantinya akan di eksekusi
#data tersebut menghilangkan baris pertama, karena pada baris pertama berisi judul
```

```python
sdata = pd.DataFrame({"Data":data})
sdata
#menampilkan data dalam bentuk data frame
```

```python
pakar = [] #import data pakar mentah dari folder
for i in range(150):
    f = open("data/pakar/"+str(i+1)+".txt", "r+", encoding='utf-8')
    lines = f.readlines()
    del lines[0]
    f.close()
    a = " ".join(lines)
    a = a.lower()
    a = a.strip()
    pakar.append(a.replace("\n", " "))
    
#memasukkan data pakar ke dalam array untuk nantinya di bandingkan dengan hasil rengkuman CLSA
#data pakar juga dilakukan preprocessing berupa case folding
#data tersebut menghilangkan baris pertama, karena baris pertama berupa judul
```

```python
spakar = pd.DataFrame({"Data Pakar":pakar})
spakar
#menampilkan data pakar dalam bentuk data frame
```

```python
prepdata = [] #preprocessing data
for i in data:
    #merubah ke huruf kecil (case folding)
    a = i.lower() 
    
    #menghapus spasi di akhir kalimat
    a = a.strip() 
    
    #memejah kalimat berdasarkan titik (data tersebut berupa array) / (tokenizing)
    a = a.split(".")
    
    #menghapus array yg kosong
    while("" in a): 
        a.remove("")
    prepdata.append(a)
    
#prepocessing data disini pake case folding dan tokenizing
#kegunaan tokenizing disini untuk nantinya dilakukan peringkasan dengan menggunakan metode
```

```python
sprepdata = pd.DataFrame(prepdata)
sprepdata.head(3)
#menampilkan data hasil prepocessing kedalam data frame
```

```python
matrixsvd = []
for i in prepdata:
    #perhitungan tfidf
    tfidf = TfidfVectorizer()
    matrix = tfidf.fit_transform(i)
    
    svd = TruncatedSVD(n_components = len(i))
    lsa = svd.fit_transform(matrix)
    
    encodmatrix=svd.components_
    matrixsvd.append(abs(encodmatrix))
#membentuk matrix SVD dari setiap data yang telah melalui tahap preprocessing
#perhitungan atau pembentukan svd diawali dengan perhitungan tfidf pada masing masing data
#perhitungan tfidf dilakukan dengan menggunakan library
```

```python
smatrixsvd = pd.DataFrame(matrixsvd[0])
smatrixsvd.head(10)
#manmpilkan 10 data teratas hasil pembentukan matrix svd pada data pertama atau data nomor 1
#data ditampilkan dalam bentuk data frame
```

```python
lsa = []
for i in matrixsvd:
    temp = []
    for j in i:
        temp.append(sum(j))
    lsa.append(temp)
#pembentukan nilai lsa pada masing masing data
```

```python
slsa = pd.DataFrame(lsa)
slsa
#menampilkan data lsa kedalam data frame
```

```python
rangkuman = []
for i in range(len(lsa)):
    temp = []
    for j in range(len(lsa[i])):
        if lsa[i][j] >= np.mean(lsa[i]):
            temp.append(prepdata[i][j])
    hasil = ".".join(temp)
    hasil = hasil.strip()
    rangkuman.append(hasil)
#menghitung dan menentukan hasil rangkuman pada metode lsa
#hasil di tentukan setelah menghitung lenght pada masing masing data
```

```python
srangkuman = pd.DataFrame({"Rangkuman":rangkuman})
srangkuman
#menampilkan hasil rangkuman kedalam dataframe
```

```python
from rouge_score import rouge_scorer
#library untuk melakukan pengujian ROUGE
```

```python
rouge1=[]
for i in range(len(rangkuman)):
    scorer = rouge_scorer.RougeScorer(['rouge1'], use_stemmer=True)
    scores = scorer.score(pakar[i], rangkuman[i])
    rouge1.append(scores["rouge1"])
hasilrouge = pd.DataFrame(rouge1)
#menguji hasil peringkasan dengan menggunakan rougeL
#pengujian dilakukan dengan menghasilkan hasil rangkuman dengan hasil pakar
```

```python
srouge = pd.DataFrame(hasilrouge)
srouge
#menampilkan hasil pengujian pada tiap masing masing data
```
